```python
import math
import matplotlib.pyplot as mp
from ipywidgets import interact
import ipywidgets as wg
e = math.e
pi = math.pi
def wavepacket(t,m,a,l,h):
    common = complex(1, 2*h*a*t/m)
    normal = 2*a/pi ** 0.25
    xvals = []
    revals = []
    imvals = []
    x = 0
    end = 10
    while x < end:
        toptop = complex(-a*x**2, l*x-h*l**2*t/2/m)
        topok = toptop/common
        value = normal*(e**topok)/(common**0.5)
        xvals.append(x)
        revals.append(value.real)
        imvals.append(value.imag)
        x += 0.05
    mp.plot(xvals,revals)
    mp.plot(xvals,imvals)
    mp.ylim(-2.8,2.8)
interact (wavepacket,t=(0.0,5.0,0.2),m=(0.01,10),a=(0,1,0.25),l=(0,3),h=(0,10));
```
